```python
from turtle import Turtle,Screen


#create Window
screen = Screen()
screen.title("Pong")
screen.setup(width=800,height=600)
screen.bgcolor("black")
screen.tracer(0)

score_1=0
score_2=0

#Paddle 1
paddle_1 = Turtle()
paddle_1.shape("square")
paddle_1.color("white")
paddle_1.shapesize(stretch_wid=5,stretch_len=1)
paddle_1.penup()
paddle_1.goto(-350,0)
paddle_1.speed(0)

#paddle 2
paddle_2 = Turtle()
paddle_2.shape("square")
paddle_2.color("white")
paddle_2.shapesize(stretch_wid=5,stretch_len=1)
paddle_2.penup()
paddle_2.goto(350,0)

 #ball
ball = Turtle()
ball.shape("circle")
ball.color("white")
ball.speed(0)
ball.penup()
ball.goto(0,0)
ball.dx = 2
ball.dy = -2 



pen = Turtle()
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("PLayer A: 0    Player B:0", align="center", font={"courier", 24,"normal"})



#function
def paddle_1_up():
    y = paddle_1.ycor() 
    y += 20
    paddle_1.sety(y)
    
    
def paddle_1_down():
    y = paddle_1.ycor() 
    y -= 20
    paddle_1.sety(y)
    
def paddle_2_up():
    y = paddle_2.ycor() 
    y += 20
    paddle_2.sety(y)
    
    
def paddle_2_down():
    y = paddle_2.ycor() 
    y -= 20
    paddle_2.sety(y)
    

#vinding
screen.listen()
screen.onkeypress(paddle_1_up, "w")
screen.onkeypress(paddle_1_down, "e")
screen.onkeypress(paddle_2_up, "Up")
screen.onkeypress(paddle_2_down, "Down")


#Main Game
while True:
    screen.update()
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    if ball.ycor()  > 290:
        ball.sety(290)
        ball.dy *= -1
    
    if ball.ycor()  < -290:
        ball.sety(-290)
        ball.dy *= -1

    if ball.xcor()  > 390:
        ball.goto(0,0)
        ball.dx *= -1
        score_1+=1
        pen.clear()
        pen.write("PLayer A: {}   Player B:{}".format(score_1,score_2), align="center", font={"courier", 24,"normal"})
        
    if ball.xcor()  < -390:
        ball.goto(0,0)
        ball.dx *= -1
        score_2+=1
        pen.clear()
        pen.write("PLayer A: {}   Player B:{}".format(score_1,score_2), align="center", font={"courier", 24,"normal"})
        
    if (ball.xcor() > 340 and ball.xcor() <350) and (ball.ycor() < paddle_2.ycor() + 40 and ball.ycor()> paddle_2.ycor() -40):
        ball.setx(340)
        ball.dx *= -1

    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_1.ycor() + 40 and ball.ycor()> paddle_1.ycor() -40):
        ball.setx(-340)
        ball.dx *= -1
```
