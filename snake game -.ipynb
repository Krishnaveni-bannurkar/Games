```python
#Snake Game -- Python

# importing libs
import pygame
import time
import random


#pygame.init() initialize all imported pygame modules.
pygame.init()

#########################################################################################################
#declarining variables along the way and displaying the game tab

#Declaring colors we use in this Game:
white = (255, 255, 255)
yellow = (255, 255, 102)
black = (0, 0, 0)
red = (213, 50, 80)
green = (0, 255, 0)
blue = (50, 153, 213)
lime = (0,255,0)


#declaring the size variables:
display_width = 600
display_height = 400

#creating a display for the game to run:
display = pygame.display.set_mode((display_width,display_height))
pygame.display.set_caption("!!Snake ~ Game!!")


clock = pygame.time.Clock()


#declaring the snake_block_size and snake_speed which we use in the main program later:
snake_block = 10 
snake_speed = 10

#########################################################################################################

# message if the snake touches the borders:
# here we are creating a surface on top of our display tab to show a message. 
font_style = pygame.font.SysFont(None , 30)
score_font = pygame.font.SysFont("comicsansms", 35)

#defining Snake

def snake(snake_block, snake_list):
    for list in snake_list:
        pygame.draw.rect(display,black,[list[0],list[1],snake_block,snake_block])



def message(msg,color):
    message_surface  = font_style.render(msg, True, color)
    display.blit(message_surface, [display_width/6,display_height/3])

#########################################################################################################
#food
#Snake movement creation in the game

def gameLoop():
    game_over = False
    game_close = False
    
    #declaring the size variables for the snake to move r,l,u,d:
    x1 = display_width/2
    y1 = display_height/2
    
    x1_change = 0
    y1_change = 0
    
    snake_list = []
    length_of_snake = 1
    
    foodx = round(random.randrange(0, display_width  - snake_block)/10.0) *10.0
    foody = round(random.randrange(0, display_height  - snake_block)/10.0) * 10.0

    while not game_over:
        while game_close == True:
            display.fill(white)
            message("You lost! Press Q-Quit or p-Play Again", lime)
            pygame.display.update() 
            time.sleep(2)
            
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_p:
                        gameLoop()            
            
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x1_change = -snake_block
                    y1_change = 0
                elif event.key == pygame.K_RIGHT:
                    x1_change = snake_block
                    y1_change = 0
                elif event.key == pygame.K_UP:
                    x1_chage = 0
                    y1_change = -snake_block
                elif event.key == pygame.K_DOWN:
                    x1_change = 0
                    y1_change = snake_block
                
        if x1 >= display_width or x1<0 or y1 >= display_height or y1 < 0: #setting the boundries
                    game_close = True 
                
        x1 += x1_change #updating the x value
        y1 += y1_change # updatig the y value
    
        display.fill(white) #displaying the screen white can change the colors
        pygame.draw.rect(display, blue, [foodx, foody, snake_block, snake_block])
        snake_head = []
        snake_head.append(x1)
        snake_head.append(y1)
        snake_list.append(snake_head)
        if len(snake_list)>length_of_snake:
            del snake_list[0]
        for list in snake_list[:-1]:
            if list == snake_head:
                game_close = True
        
        snake(snake_block,snake_list)
        pygame.display.update() #after moving the snake update the display
        
        
        if x1 == foodx and y1 == foody:
            foodx = round(random.randrange(0, display_width - snake_block) / 10.0) * 10.0
            foody = round(random.randrange(0, display_height - snake_block) / 10.0) * 10.0
            length_of_snake += 1
        clock.tick(snake_speed)
    
    
    pygame.quit()
    quit()

gameLoop()
```
