```python
import random

MAX_LINES = 3
MAX_BET = 1000
MIN_BET = 1

ROWS= 3
COLS= 3

Symbol_count ={
    "A" : 2,
    "B" : 4,
    "C" : 6,
    "D" : 8
}


Symbol_value ={
    "A" : 5,
    "B" : 4,
    "C" : 3,
    "D" : 2
}


def check_winnings(columns, lines, values, bet):
    winnings = 0
    winning_lines = []
    for line in range(lines):
        symbol = columns[0][line]
        for column in columns:
            symbol_to_check = column[line]
            if symbol != symbol_to_check:
                break
        else: 
            winnings += values[symbol] * bet
            winning_lines.append(line +1)
    return winnings , winning_lines

def get_slot_machine_spin(rows, cols, symbols):
    all_symbols=[]
    for symbol, symbol_count in symbols.items():
        for _ in range(symbol_count):
            all_symbols.append(symbol)
#result shoul be ['A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C', 'C', 'D', 'D', 'D', 'D', 'D', 'D', 'D', 'D'] 

    
    columns = [] # [[],[],[]] result set would look like this
    for _ in range(cols): #this will loop the num of column times and create a list
        column = []
        current_symbols = all_symbols[:]
        for _ in range(rows):
            value = random.choice(current_symbols)
            current_symbols.remove(value)
            column.append(value)
            
        columns.append(column)
    
    return columns

def print_slot_machine(columns):
    for row in range(len(columns[0])):
        for i , column in enumerate(columns):
            if i != len(columns) - 1:
                print(column[row], end= " | ")
            else:
                print(column[row], end= "")
        print()
                
#taking the amount input
def deposit():
    while True:
        amount = input("Please enter the deposit Amount: $")
        if amount.isdigit():
            amount = int(amount)
            if amount > 0:
                break
            else: 
                print("Amount must be greater than 0 ")
        else: 
            print("please enter a Number(digit)")
    return amount


#number of lines to bet
def get_number_of_lines():
    while True:
        lines = input("Enter the number of lines you want to bet on (1-" + str(MAX_LINES) + ")? ")
        if lines.isdigit():
            lines = int(lines)
            if 1 <= lines <= MAX_LINES:
                break
            else : 
                print("Enter valind bet")
        else:
            ("Please enter a number")
    return lines


#amount that gamesr wants to bet, he can use entire amount or a fraction so we are defining this function bet
def get_bet():
    while True:
        bet_amount = input("Please enter amount you want to bet on each line: $")
        if bet_amount.isdigit():
            bet_amount = int(bet_amount)
            if MIN_BET <= bet_amount <= MAX_BET:
                break;
            else: 
                print(f"Please enter a valid bet between ${MIN_BET} - ${MAX_BET}")
        else:
            print("PLease enter a Number/Digit")
    return bet_amount


def spin(balance):
    lines = get_number_of_lines()
    while True:
        bet = get_bet()
        total_bet = lines * bet
        if total_bet >= balance:
            print(f"You do not have enough to bet, your current balance is ${balance} and you are betting: ${total_bet}")
        else:
            break
    print(f"You are betting ${bet} each on {lines} lines. Total Bet is ${total_bet}")
    
    slot = get_slot_machine_spin(ROWS, COLS, Symbol_count)
    print(print_slot_machine(slot))
    winnings , winning_lines = check_winnings(slot, lines, Symbol_value, bet)
    print(f"You won {winnings}")
    print(f"You won on", *winning_lines)
    return winnings - total_bet
    
def main():
    balance = deposit()
    while True:
        print(f"current balance is ${balance}")
        result = input("Press enter to play (q to Quit)")
        if result == "q":
            break
        balance += spin(balance)
    print(f"you are lef with ${balance}")
main()
```
